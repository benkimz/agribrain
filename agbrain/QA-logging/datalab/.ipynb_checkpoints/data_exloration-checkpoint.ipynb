```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataframe = pd.read_csv("article_summary.csv")
```

```python
dataframe.head()
```

```python
dataframe.describe()
```

```python
list(dataframe["category"].unique())
```

```python
dataframe["article_chars"] = dataframe["article"].map(lambda article: len(article))
dataframe["summary_chars"] = dataframe["summary"].map(lambda summary: len(summary))
```

```python
dataframe.head(10)
```

```python
dataframe.describe()
```

```python
dataframe.hist(bins=60, figsize=[20, 10])
```

```python
# want to view extremely long summaries
long_summaries = dataframe[dataframe["summary_chars"] > 2500]
long_summaries
```

```python
long_summaries[["article_chars", "summary_chars"]].plot(kind="scatter", x="article_chars", y="summary_chars")
```
